```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Use glob to find all files starting with 'cluster1_16TB' in the current directory
files = glob.glob('cluster*')
files.sort()
# Loop through each file
for file_path in files:
    print(f"\n\n\n\n {file_path}:")
    df = pd.read_csv(file_path)
    
    # Function to convert bytes to KB or MB
    def convert_size(size_bytes):
        if size_bytes < 1024 * 1024:  # Less than 1MB
            return f"{size_bytes / 1024:.2f} KB"
        else:
            return f"{size_bytes / (1024 * 1024):.2f} MB"
    
    # Function to analyze and print distribution
    def analyze_distribution(column, name):
        counts = df[column].value_counts()
        total = len(df)
        print(f"\n{name} Distribution:")
        for value, count in counts.items():
            percentage = (count / total) * 100
            print(f"{value}: {count} ({percentage:.2f}%)")
    
    # Function to analyze I/O sizes
    def analyze_io_size(column, name):
        sizes = df[column]
        print(f"\n{name} Analysis:")
        print(f"Distinct sizes: {sizes.nunique()}")
        print(f"Lowest size: {sizes.min()} bytes ({convert_size(sizes.min())})")
        print(f"Highest size: {sizes.max()} bytes ({convert_size(sizes.max())})")
        
        print("\nTop 10 sizes (most frequent):")
        top_10 = sizes.value_counts().nlargest(10)
        for size, count in top_10.items():
            percentage = (count / len(sizes)) * 100
            print(f"{size} bytes ({convert_size(size)}): {count} ({percentage:.2f}%)")
    
    # Perform analysis
    print(f"Number of traces: {len(df)}")
    print(f"Number of distinct files: {df['filename'].nunique()}")
    
    analyze_distribution('io_zone', 'IO Zone')
    analyze_distribution('op_type', 'Operation Type')
    analyze_distribution('service_class', 'Service Class')
    analyze_distribution('from_flash_cache', 'From Flash Cache')
    analyze_distribution('cache_hit', 'Cache Hit')
    
    analyze_io_size('request_io_size_bytes', 'Request I/O Size')
    analyze_io_size('response_io_size_bytes', 'Response I/O Size')
    analyze_io_size('disk_io_size_bytes', 'Disk I/O Size')
    
    # Create a vertical bar chart for main categories
    categories = ['io_zone', 'op_type', 'service_class', 'from_flash_cache', 'cache_hit']
    fig, ax = plt.subplots(figsize=(12, 6))
    
    x_offset = 0
    bar_width = 0.8
    space_between = 0.4
    colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']
    
    for i, category in enumerate(categories):
        data = df[category].value_counts(normalize=True) * 100
        x = [x_offset + j for j in range(len(data))]
        ax.bar(x, data.values, width=bar_width, label=category, color=colors[i % len(colors)])
        
        for j, (value, percentage) in enumerate(data.items()):
            ax.text(x[j], percentage, f'{value}\n{percentage:.1f}%', ha='center', va='bottom', rotation=90)
        
        x_offset += len(data) + space_between
    
    ax.set_xticks([])
    ax.set_ylabel('Percentage')
    ax.set_title('Distribution of Categories')
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
    
    plt.tight_layout()
    plt.show()
```
